```python
import pandas as pd
from scipy.stats import chi2_contingency, mannwhitneyu
```

# Hypotéza 1
Na silnicích první třídy se při nehodách umíralo se stejnou pravděpodobností jako na silnicích třetí třídy.

Načteme data a zobrazíme tabulku

```python
df = pd.read_pickle("accidents.pkl.gz")
df
```

Rozhodneme, u kterých nehod došlo k fatálním následkům a omezíme rozsah na silnice 1. a 3. třídy

```python
df2 = df[(df["p36"].isin([1, 3]))].copy()
df2["death"] = df2["p13a"] > 0
df2
```

Vytvoříme kontingenční tabulku, která zobrazuje počet smrtelných nehod a nehod bez umrtí na silnicích 1. a 3. třídy.

```python
ct = pd.crosstab(df2['death'], df2['p36'])
ct
```

Nyní s pomocí **$\chi^2$ chí-kvadrát testu** ověříme, jestli existuje mezi proměnnými statisticky významná korelace.

```python
result = chi2_contingency(ct)
result[1]
```

Hodnota $3.54\cdot10^{-29}$ je mnohem menší než $0.05$. Proto zamítneme nulovou hypotézu.

Očekávané frekvence:

```python
result[3]
```

```python
ct - result[3]
```

Z výstupu $\chi^2$ chí-kvadrát testu je také vidět, že nehody na silnicích 1. třídy vedly častěji ke smrtelným zraněním.

# Hypotéza 2

Při nehodách vozidel značky Škoda je škoda na vozidle nižší než při nehodách vozidel Audi.

Nejprve vyfiltrujeme vozy Audi (2) a Škoda (39) a vyčíslení opravy.

```python
# 2 == Audi, 39 == Skoda
df3 = df[df["p45a"].isin([2, 39])].copy()
df3 = df3[["p45a", "p53"]]
df3
```

Dále vypočítáme porovnání dvou konfigurací (Audi a Škoda), porovnáme mediány a spočítáme výsledek MWU-testu.

```python
df4 = df3.set_index("p45a")
a = df4.loc[2, "p53"]
b = df4.loc[39, "p53"]

print(a.median(), b.median())
res = mannwhitneyu(a, b)
res
```

Z výsledku je vidět, že `pvalue` je mnohem menší než stanovená hodnota `0.05`. To znamená, že jsme testem vyvrátili podobnost v ceně opravy po nehodě u aut značky Škoda a Audi. Tomuto odpovídají i porovnané mediány, ze kterých je vidět, že opravy aut Audi jsou dražší.
